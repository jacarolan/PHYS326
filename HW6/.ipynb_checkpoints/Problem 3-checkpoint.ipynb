```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
from pylab import polyfit
```

# Our system:
    dv/dt = -0.1v - sin(x) + γ*cos(ωt)*cos(x)
    dx/dt = v

```python
def d(x, v, t, w, l):
    return v, -0.1 * v - np.sin(x) + l * np.cos(w * t) * np.cos(x)    # returns dx/dt, dv/dt
```

# (a) Solve system with quiescent initial conditions

```python
l = 0.6
w = 0.9

x, v = 0, 0
dt = 0.2        # Time step size
sim_length = 500  # How long to simulate for

vs = [v]     # These are arrays right now so appending is easy


# We will use classical fourth order runge kutta
for t in np.linspace(0, sim_length - dt, int(sim_length / dt)):
    # Compute approximate steps
    k_1 = [dt * i for i in d(x, v, t, w, l)]
    k_2 = [dt * i for i in d(x + k_1[0] / 2, v + k_1[1] / 2, t + dt / 2, w, l)]
    k_3 = [dt * i for i in d(x + k_2[0] / 2, v + k_2[1] / 2, t + dt / 2, w, l)]
    k_4 = [dt * i for i in d(x + k_3[0], v + k_3[1], t + dt, w, l)]
    
    # Take the step based on weighted average of possible steps
    x += (k_1[0] + 2 * k_2[0] + 2 * k_3[0] + k_4[0]) / 6
    v += (k_1[1] + 2 * k_2[1] + 2 * k_3[1] + k_4[1]) / 6
    vs.append(v)
    
vs = np.array(vs)
plt.title("Velocity over time")
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.plot(np.linspace(0, sim_length, 1 + int(sim_length / dt)), vs)
```

## Plot a reasonable amount of cycles

```python
plt.title("Velocity over time")
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.plot(np.linspace(0, sim_length, 1 + int(sim_length / dt))[-155:-5], vs[-155:-5])
```

The attractor will trace a closed, non-self-intersecting curve in phase space, because the velocity only changes direction at a peak or a trough. This means the attractor is period 1. The graph indicates that the attractor is obviously not sinusoidal.

# (b) Solving with new parameters

```python
l = 1.22
w = 0.9

x, v = 0, 0
dt = 2 * np.pi / (40 * w)        # Time step size
sim_length = 20000  # How long to simulate for

vs = [v]     # These are arrays right now so appending is easy
xs = [x]


# We will use classical fourth order runge kutta
for t in np.linspace(0, sim_length - dt, int(sim_length / dt)):
    # Compute approximate steps
    k_1 = [dt * i for i in d(x, v, t, w, l)]
    k_2 = [dt * i for i in d(x + k_1[0] / 2, v + k_1[1] / 2, t + dt / 2, w, l)]
    k_3 = [dt * i for i in d(x + k_2[0] / 2, v + k_2[1] / 2, t + dt / 2, w, l)]
    k_4 = [dt * i for i in d(x + k_3[0], v + k_3[1], t + dt, w, l)]
    
    # Take the step based on weighted average of possible steps
    x += (k_1[0] + 2 * k_2[0] + 2 * k_3[0] + k_4[0]) / 6
    v += (k_1[1] + 2 * k_2[1] + 2 * k_3[1] + k_4[1]) / 6
    vs.append(v)
    xs.append(x)
    
vs = np.array(vs)
xs = np.array(xs)


plt.title("Velocity over time")
plt.xlabel("Time")
plt.ylabel("Velocity")

# Make the plot readable by plotting part of the tail end
plt.plot(np.linspace(0, sim_length, 1 + int(sim_length / dt))[-3000:], vs[-3000:])
```

This is chaotic behaviour over extended period of time.

## Plot poincare section

```python
nvals = int(sim_length / (dt * 40))

x = np.array([xs[40 * i] for i in range(nvals)]) % (2 * np.pi)  # Positions
y = np.array([vs[40 * i] for i in range(nvals)])  # Velocities

plt.plot(x, y, ls='', marker=',')
plt.title("Poincare section")
plt.xlabel("Position mod 2 * pi")
plt.ylabel("Velocity")
plt.show()
```

# (c) Comparing similar initial conditions

```python
l = 1.22
w = 0.9

x, v = 0, 0
dt = 2 * np.pi / (40 * w)        # Time step size
sim_length = 100  # How long to simulate for

vs = [v]     # These are arrays right now so appending is easy
xs = [x]


# We will use classical fourth order runge kutta
for t in np.linspace(0, sim_length - dt, int(sim_length / dt)):
    # Compute approximate steps
    k_1 = [dt * i for i in d(x, v, t, w, l)]
    k_2 = [dt * i for i in d(x + k_1[0] / 2, v + k_1[1] / 2, t + dt / 2, w, l)]
    k_3 = [dt * i for i in d(x + k_2[0] / 2, v + k_2[1] / 2, t + dt / 2, w, l)]
    k_4 = [dt * i for i in d(x + k_3[0], v + k_3[1], t + dt, w, l)]
    
    # Take the step based on weighted average of possible steps
    x += (k_1[0] + 2 * k_2[0] + 2 * k_3[0] + k_4[0]) / 6
    v += (k_1[1] + 2 * k_2[1] + 2 * k_3[1] + k_4[1]) / 6
    vs.append(v)
    xs.append(x)
    
vs = np.array(vs)
xs = np.array(xs)

plt.title("Velocity over time")
plt.xlabel("Time")
plt.ylabel("Velocity")
plt.plot(np.linspace(0, sim_length, 1 + int(sim_length / dt)), vs)

x2, v2 = 0, 1e-5

v2s = [v2]     # These are arrays right now so appending is easy
x2s = [x2]


# We will use classical fourth order runge kutta
for t in np.linspace(0, sim_length - dt, int(sim_length / dt)):
    # Compute approximate steps
    k_1 = [dt * i for i in d(x2, v2, t, w, l)]
    k_2 = [dt * i for i in d(x2 + k_1[0] / 2, v2 + k_1[1] / 2, t + dt / 2, w, l)]
    k_3 = [dt * i for i in d(x2 + k_2[0] / 2, v2 + k_2[1] / 2, t + dt / 2, w, l)]
    k_4 = [dt * i for i in d(x2 + k_3[0], v2 + k_3[1], t + dt, w, l)]
    
    # Take the step based on weighted average of possible steps
    x2 += (k_1[0] + 2 * k_2[0] + 2 * k_3[0] + k_4[0]) / 6
    v2 += (k_1[1] + 2 * k_2[1] + 2 * k_3[1] + k_4[1]) / 6
    v2s.append(v2)
    x2s.append(x2)
    
v2s = np.array(v2s)
x2s = np.array(x2s)

plt.plot(np.linspace(0, sim_length, 1 + int(sim_length / dt)), v2s)
plt.legend(["Quiescent Conditions", "Perturbed Conditions"])
plt.show()
```

## Plot logarithmic divergence

```python
# Use only the first ~50 time units to avoid additional chaos
x = np.linspace(0, 50, int(50 / dt) - 2)
diff_original = np.abs(xs[:int(50/dt)] - x2s[:int(50/dt)])
diff = []
for i in range(1, len(diff_original) - 1):
    diff.append(sum(diff_original[i-1:i+2]) / 3) # Smooth raw distance to avoid having 0's
diff = np.array(diff)

plt.plot(x, np.log(diff))
plt.show()
```

## Estimate Lyapunov Exponent

```python
m,b = polyfit(x, np.log(diff), 1)
print("Lyapunov exponent:", m)
```
