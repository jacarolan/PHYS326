# Problem 2

```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

# Our system:
    dv/dt = -0.1v - 4x + cos(t)
    dx/dt = v

```python
def d(x, v, t):
    return v, -0.1 * v - 4 * x + np.cos(t)    # returns dx/dt, dv/dt
```

# Solve system with quiescent initial conditions

```python
x, v = 0, 0
dt = 0.001        # Time step size
sim_length = 100  # How long to simulate for

xs = [x]     # These are arrays right now so appending is easy


# We will use classical fourth order runge kutta
for t in np.linspace(0, sim_length - dt, int(sim_length / dt)):
    # Compute approximate steps
    k_1 = [dt * i for i in d(x, v, t)]
    k_2 = [dt * i for i in d(x + k_1[0] / 2, v + k_1[1] / 2, t + dt / 2)]
    k_3 = [dt * i for i in d(x + k_2[0] / 2, v + k_2[1] / 2, t + dt / 2)]
    k_4 = [dt * i for i in d(x + k_3[0], v + k_3[1], t + dt)]
    
    # Take the step based on weighted average of possible steps
    x += (k_1[0] + 2 * k_2[0] + 2 * k_3[0] + k_4[0]) / 6
    v += (k_1[1] + 2 * k_2[1] + 2 * k_3[1] + k_4[1]) / 6
    xs.append(x)
    
xs = np.array(xs)

plt.title("Position versus time")
plt.xlabel("Time")
plt.ylabel("Position")
plt.plot(np.linspace(0, sim_length, 1 + int(sim_length / dt)), xs)
```

The solution starts quiescent, there is a transient, which fades away into an oscillatory solution of approximately (1/3)cos(t)
