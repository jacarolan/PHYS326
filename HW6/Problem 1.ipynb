# Problem 1

```python
import numpy as np
import numpy.linalg as la
import matplotlib.pyplot as plt
```

```python
def h(x, y):
    return (y + 1 - 1.4 * x ** 2, 0.3 * x)
```

# Plot iterations using the starting point (0, 0)

```python
x = np.zeros(10000)
y = np.zeros(10000)
for i in range(9999):
    x[i + 1], y[i + 1] = h(x[i], y[i])

plt.plot(x, y, ls='', marker=',')
plt.title("Iterations starting at (0, 0)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# Plot iterations using the starting point (0.1, 0.1)

```python
x = np.zeros(10000)
y = np.zeros(10000)
x[0], y[0] = 0.1, 0.1
for i in range(9999):
    x[i + 1], y[i + 1] = h(x[i], y[i])

plt.plot(x, y, ls='', marker=',')

plt.title("Iterations starting at (0.1, 0.1)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

It appears that this folded horse shoe shape is an attractor of the henon map, as different starting condtions tend to all wind up along it.
