# Definitions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x, alpha):
    return alpha * np.sin(x * np.pi)

def f(x, alpha, iters):
    for i in range(iters):
        x = g(x, alpha)
    return x
```

# Plot Bifurcation diagram

```python
P=1-np.geomspace(0.01,1,10000)
X = []
Y = []
m = 0
for u in P:
    X.append(u)
    m = f(np.random.random(), u, 2000)
    Y.append(m)
plt.figure(figsize=(20, 8))
plt.plot(X, Y, ls='', marker=',')
plt.show()
```

This is similar to the log map bifurcation graph shown in class.

# Finding Lyapunov Exponent

```python
epsilon = 1e-9

runs = 2000
iters = 50

lyapunov = 0

for i in range(runs):
    v = np.random.random() * (1 - epsilon)
    v_prime = v + epsilon
    diff = np.zeros(iters)
    for j in range(iters):
        diff[j] = np.abs(v - v_prime)
        v = g(v, 0.9)
        v_prime = g(v_prime, 0.9)
    inf = np.polyfit(np.arange(iters), np.log(diff), 1)
    lyapunov += inf[0]

lyapunov /= runs

print(lyapunov)
```
