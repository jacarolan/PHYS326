# Definitions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g(x, alpha):
    return 2 * alpha * x if x < 0.5 else 2 * alpha * (1-x)

def f(x, alpha, iters):
    for i in range(iters):
        x = g(x, alpha)
    return x
```

# Part (a)

```python
X = []
Y = []

x = 0.2
alpha = 0.4
X.append(0)
Y.append(x)
for i in range(20):
    x = g(x, alpha)
    Y.append(x)
    X.append(i + 1)

plt.scatter(X, Y)
plt.title("Plot of sequence with alpha = 0.4")
plt.ylabel("x value")
plt.xlabel("Iteration")
plt.show()

X = []
Y = []

x = 0.2
alpha = 0.7
X.append(0)
Y.append(x)
for i in range(20):
    x = g(x, alpha)
    Y.append(x)
    X.append(i + 1)

plt.title("Plot of sequence with alpha = 0.7")
plt.ylabel("x value")
plt.xlabel("Iteration")
plt.scatter(X, Y)
plt.show()
```

alpha = 0.7 is chaotic, but alpha = 0.4 is not chaotic.

# Part (b)

```python
P=1-np.geomspace(0.01,1,10000)
X = []
Y = []
m = 0
for u in P:
    X.append(u)
    m = f(np.random.random(), u, 2000)
    Y.append(m)
plt.figure(figsize=(20, 8))
plt.plot(X, Y, ls='', marker=',')
plt.show()
```

This map appears to be stable for alpha less than 0.5, but very quickly becomes chaotic afterwards. There is a pull towards 0 for alpha less than 0.5, which makes sense because x<sub>n+1</sub> will be less than x if alpha is less than 0.5.There is a breif period of apparent stability with an attractor cycle until about alpha=0.6, at which point total chaos begins to set in.
